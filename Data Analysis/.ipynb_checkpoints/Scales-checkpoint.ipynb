```python
import pandas
from glob import glob
```

```python
files = glob('./MIDIfiles/Scales/csv/*.csv')
```

```python
df = pandas.read_csv(files[0], header = None , names=['Onset (Beats)','Duration (Beats)','MIDI Channel','MIDI Pitch','Velocity','Onset (Sec)','Duration (sec)'])
```

```python
notes = df['MIDI Pitch'].tolist()
```

```python
notes
```

```python
correctNotes = pandas.read_csv('correctScales.csv')
correctNotes
```

```python
score = []
cmajor = correctNotes[correctNotes.columns[0]].tolist()
c = 1

while c < 12: #for the 12 scales
    temp = 0
    cstart = c
    for n in range(len(notes)-14):
        if (notes[n] == cmajor[0]) and (notes[n+7] == cmajor[7]) and (notes[n+14] == cmajor[14]):
            for p in range(15):
                if notes[n + p] != cmajor[p]:
                    temp = temp + 1
            c = c + 1
            if c > 12:
                break
    if c == cstart:
        c = c + 1
        temp = 15
    score.append(temp)
            
```
